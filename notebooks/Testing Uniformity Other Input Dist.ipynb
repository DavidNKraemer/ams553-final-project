```python
def chiTest(n = 1000, d = 200):
    duniform_drones = np.random.randint(1, 1000-1, size=1000)
    bins = np.linspace(0, 1000-1, n//d + 1)
    d, _ = np.histogram(duniform_drones, bins)
    return stats.chisquare(d)[-1] # assumes all bins equally likely
```
